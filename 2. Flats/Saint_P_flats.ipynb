# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

Описание данных<br>
airports_nearest — расстояние до ближайшего аэропорта в метрах (м)<br>
balcony — число балконов<br>
ceiling_height — высота потолков (м)<br>
cityCenters_nearest — расстояние до центра города (м)<br>
days_exposition — сколько дней было размещено объявление (от публикации до снятия)<br>
first_day_exposition — дата публикации<br>
floor — этаж<br>
floors_total — всего этажей в доме<br>
is_apartment — апартаменты (булев тип)<br>
kitchen_area — площадь кухни в квадратных метрах (м²)<br>
last_price — цена на момент снятия с публикации<br>
living_area — жилая площадь в квадратных метрах(м²)<br>
locality_name — название населённого пункта<br>
open_plan — свободная планировка (булев тип)<br>
parks_around3000 — число парков в радиусе 3 км<br>
parks_nearest — расстояние до ближайшего парка (м)<br>
ponds_around3000 — число водоёмов в радиусе 3 км<br>
ponds_nearest — расстояние до ближайшего водоёма (м)<br>
rooms — число комнат<br>
studio — квартира-студия (булев тип)<br>
total_area — площадь квартиры в квадратных метрах (м²)<br>
total_images — число фотографий квартиры в объявлении<br>
Пояснение: апартаменты — это нежилые помещения, не относящиеся к жилому фонду, но имеющие необходимые условия для проживания

Думаю, если человек не указал в объявлении количество балконов, то скорее всего их просто нет. И в данном случае можно заменить пропущенные значения на 0:

```python
df['balcony'] = df['balcony'].fillna(0)
```

```python
df[df['floors_total'].isna()]
```

Заменим кол-во этажей на медиану в зависимости от населенного пункта (будем считать, что застройка в населенных пунктах типовая):

```python
df['floors_total'] = df.groupby('locality_name')['floors_total'].apply(lambda x: x.fillna(x.median()))
```

```python
df[df['floors_total'].isna()]
```

Там, где нет названия населенного пункта заменим кол-во этажей в доме на медианное значение этажа всего датафрейма:

```python
df['floors_total'] = df['floors_total'].fillna(df['floors_total'].median())
```

```python
df[df['floors_total'] < df.floor]
```

Осталось несколько квартир, где после замены на медиану этаж оказывается больше, чем всего этажей в доме. Думаю, тут можно считать, что квартира находится на последнем этаже:

```python
df['floors_total'] = df.apply(lambda x: x['floor'] if x['floors_total'] < x['floor'] else x['floors_total'], axis = 1 )
```

```python
df[df['floors_total'] < df.floor]
```

```python
df['floors_total'].describe()
```

Двигаемся дальше:

```python
df[df['living_area'].isna()]
```

Посчитаем жилую площадь как разницу между общей площадью и площадью кухни. Да, это, конечно, будет не точно, потому что есть ванные, туалеты, балконы, но в целом это будет больше похоже на правду, чем совсем ничего.

```python
df['living_area'] = df['living_area'].fillna(df['total_area'] - df['kitchen_area'])
```

```python
df[df['living_area'].isna()]
```

Все еще много пропусков в жилой площади. Посмотрим, как обычно соотносится общая площадь и жилая, найдем медиану и используем ее как коэффициент:

```python
ratio_area = df['living_area']/df['total_area']
ratio_median = ratio_area.median() 
```

```python
df['total_area'].isna().sum()
```

Т.к. в total_area пропущенных значений нет, то можно заменить пропущенные значения в living_area:

```python
df['living_area'] = df.apply(lambda x: x['total_area']*ratio_median if pd.isna(x['living_area']) else x['living_area'], axis =1)
df['living_area'].isna().sum()
```

```python
df[df['kitchen_area'].isna()]
```

Предлагаю поступить так же, как и с жилой площадью. Проверим соотношение общей площади и площади кухни, найдем медиану, используем как коэффициент:

```python
ratio_kitchen = df['kitchen_area']/df['total_area']
ratio_kitchen_median = ratio_kitchen.median() 
```

```python
df['kitchen_area'] = df.apply(lambda x: x['total_area']*ratio_kitchen_median if pd.isna(x['kitchen_area']) else x['kitchen_area'], axis =1)
df['kitchen_area'].isna().sum()
```

```python
df.info()
```

Пропусков все меньше! Кажется, определение населенного пункта по имеющимся данным - это гадание на кофейной гуще, поэтому думаю можно пропущенные значения заменить на "Нет данных":

```python
df['locality_name'] = df['locality_name'].fillna("Нет данных")
```

```python
df['days_exposition'].isna().sum()
```

Пропуски в days_exposition можно объяснить тем, что объявление еще не было снято. Заменим пропуски на 0:

```python
df['days_exposition'] = df['days_exposition'].fillna(0).astype('int')
```

Чудно! Двигаемся дальше. 
Приведем данные с датой к нужному формату:

```python
 df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

Картографические данные трогать не будем - если значение пропущено, то , скорее всего, данных действительно нет. Рядом нет парков или водоемов. 

```python
print(df['parks_around3000'].isna().mean())
print(df['ponds_around3000'].isna().mean())
```

В столбцах с числом парков и водоемов заменим пропуски на 0:

```python
df['parks_around3000'] = df['parks_around3000'].fillna(0)
df['ponds_around3000'] = df['ponds_around3000'].fillna(0)
```

```python
df.info()
```

Вывод: Просмотрели данные, заменили пропущенные значения.

## Расчёты и добавление результатов в таблицу

1. Цена квадратного метра

```python
df['price_per_square_meter'] = df['last_price']/df['total_area']
```

2. День недели, месяц и год публикации объявления

```python
df['day_of_week'] = df['first_day_exposition'].dt.dayofweek
```

```python
df['month'] = df['first_day_exposition'].dt.month
```

```python
df['year'] = df['first_day_exposition'].dt.year
```

```python
df.info()
```

3. Этаж квартиры; варианты — первый, последний, другой

```python
def flat_floor(row):
    if row['floor'] == 1:
        return 'Первый'
    elif row['floor'] == row['floors_total']:
        return 'Последний'
    else:
        return 'Другой'
        
df['flat_floor'] = df.apply(lambda x: flat_floor(x), axis = 1)
```

4. Соотношение жилой и общей площади, а также отношение площади кухни к общей

```python
df['living_vs_total'] = df['living_area'] / df['total_area']
df['kitchen_vs_total'] = df['kitchen_area'] / df['total_area']
```
